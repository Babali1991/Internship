```python
# Find the IQR (inter quantitle range) to identify outliers

# 1st quantile
q1 = data.quantile(0.25)

# 3rd quantile
q3 = data.quantile(0.25)

# IQR
iqr = q3-q1
```

```python
q1
```

Work on outlier detection.


Outlier detection formula
Higher side = q3 +(1.5*IQR)
Lower  side = q1 - (1.5*IQR)

```python
 # Valditing outlier  
fa_high = q3['fixed acidity'] + (1.5 * iqr['fixed acidity'])
fa_high
```

```python
# Check the indexex which have higher values
np_index = np.where(data['fixed acidity'] > fa_high)
np_index
```

```python
# Drop the index which we found in the above cell
data = data.drop(data.index[np_index])
data.shape
```

```python
data.reset_index()
```
